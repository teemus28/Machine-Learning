```python
#we can add title also
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("1st plot")

x2=np.array([8,4,2,6])
y2=np.array([8,3,5,1])

plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.title("2nd plot")
```

```python
#add supertitle
plt.subplot(1,2,1)
plt.plot(x,y)

x2=np.array([8,4,2,6])
y2=np.array([8,3,5,1])

plt.subplot(1,2,2)
plt.plot(x2,y2)

plt.suptitle("main plot")
```
