```python
# 1.1 EDA  numrical and categorical - Multivariant Analysis

import pandas as pd
import seaborn as sbn
```

```python
# Dataset

tips = sbn.load_dataset('tips') # restaurant -> $100 -> Table were full , Big party (4 - 10 ) ->  100 - 10  20 - 5
tips

titanic = pd.read_csv('/content/titanic-dataset.csv') # titanic
titanic
titanic.shape

iris = sbn.load_dataset('iris') # flower
iris

sbn.get_dataset_names()
```

```python
# 1.2 colums. - numercic and categorical data
# Scatterplot Numeric - Numeric

sbn.scatterplot(x=tips['total_bill'], y= tips['tip'], hue=tips['sex'], style=tips['smoker'], size=tips['size'])
```

```python
tips
```

```python
# Numerical and Categorical
sbn.barplot(x=titanic['Pclass'],y=titanic['Fare'], hue=titanic['Sex'])
```

```python
sbn.boxplot(x=titanic['Sex'],y=titanic['Age'], hue=titanic['Survived'])
```

```python
# Distplot (Numerical - Catergorical)

sbn.distplot(titanic[titanic['Survived']==0]['Age'], hist=False, color = 'red')
sbn.distplot(titanic[titanic['Survived']==1]['Age'], hist=False , color = 'green')
```
